```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page where the GME revenue data is located
url = 'THE_URL_CONTAINING_GME_REVENUE_DATA'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table that contains the GME revenue data. 
# Adjust the selector as needed to correctly target the table in your specific case.
table = soup.find('table', attrs={'class': 'YOUR_TABLE_CLASS_OR_ID'})

# Initialize a list to store your data
data = []

# Assuming the first row is the header
headers = [header.text for header in table.find_all('th')]

# Process table rows, adjust as necessary for your specific table structure
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) > 0:  # This checks to make sure you are not reading a blank row
        data_row = [cell.text.strip() for cell in cols]
        data.append(data_row)

# Create a DataFrame with the extracted data
gme_revenue = pd.DataFrame(data, columns=headers)

# Display the last five rows of the DataFrame
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price ($)')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
